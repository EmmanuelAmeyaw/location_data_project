<p style="margin-bottom: 0in; line-height: 100%;" align="justify"><span style="font-size: large;">From the table above, we see that the best models are 1. <strong>AUTO-ENCODER + KMEANS </strong><span style="font-weight: normal;">and 2. </span><strong>AUTO-ENCODING + DBSCAN. </strong><span style="font-weight: normal;">These two models give similar results in terms of the cluster properties of each cluster. Thus, for the two models, the distinct clusters can be classified as lower class, upper class and middle class according to mean household expenditure in each cluster. The advantage of </span><span style="font-weight: normal;">the</span> <strong>AUTO-ENCODING + DBSCAN</strong> <span style="font-weight: normal;">model </span><span style="font-weight: normal;">is that, </span><span style="font-weight: normal;">it is more robust to outliers while the </span><strong>AUTO-ENCODING + </strong><strong>KMEANS</strong><span style="font-weight: normal;"> is not. Though outliers were found in our dataset, the results above indicate that the outliers have a less significant effect on the the kmeans clustering after applying auto-encoding. </span></span></p>
